```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Generating Training Set

```python
N_train = 10
X_train = np.random.uniform(0, 1, N_train)
t_train = []

for data in X_train:
    epsilon = np.random.normal(0, 0.3)
    t_train.append(np.sin(2*np.pi*data) + epsilon)
plt.scatter(X_train, t_train)
```

#### Generating Test Set

```python
N_test = 100
X_test = np.random.uniform(0, 1, N_test)
t_test = []

for data in X_test:
    epsilon = np.random.normal(0, 0.3)
    t_test.append(np.sin(2*np.pi*data) + epsilon)
plt.scatter(X_test, t_test)
```
